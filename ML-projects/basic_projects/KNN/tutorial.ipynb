```python
# All the required import statements
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math as mt
```

```python
x = []
y = []
for i in range(100):
    x.append(i)
    y.append(i**2)

print(x)
print(y)

plt.plot(x,y)
plt.show()
```

```python
x = []
y = []

for i in range(100):
    x.append(mt.sin(i))
    y.append(mt.cos(i))
    
plt.plot(x,y)
plt.show()
```

```python
x = []
y  =[]

for i in range(1,100):
    x.append(1 / mt.cos(i))
    y.append(1 / mt.sin(i))
    
plt.plot(x,y)
plt.show()
```

```python
x = []
y  =[]

for i in range(1,100):
    x.append(mt.tan(i))
    y.append(1 / mt.tan(i))
    
plt.plot(x,y)
plt.show()
```

```python
x = []
y = []

for i in range(1000):
    x.append(i)
    y.append(mt.pow(i,5))

plt.plot(x,y)
plt.show()
    
```

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
# Importing the data set
data = pd.read_csv("C:\\Users\\risha\\OneDrive\\Desktop\\Machine_Learning\\lab\\NN-lab\\studentscore.csv")
print(data)
```

```python
#splitting data into training and testing set
X_train = data.iloc[:20,0].values
y_train = data.iloc[:20,1].values
X_test = data.iloc[21:,0].values
y_test= data.iloc[21:,1].values
X_train = X_train.reshape(20,1)
y_train = y_train.reshape(20,1)
X_test = X_test.reshape(4,1)
y_test = y_test.reshape(4,1)
# print(X_test)
# print(X_train)
# print(y_train)
```

```python
reg = linear_model.LinearRegression()
reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)
y_pred_2 = reg.predict(X_train)
```

```python
# Printing the coefficients and errors
print("Coefficients:\t {}".format(reg.coef_))
print("Mean absolut error: \t {}".format(mean_absolute_error(y_test,y_pred)))
print("Mean squared error: \t {}".format(mean_squared_error(y_test,y_pred)))
print("R2 error:\t {}".format(r2_score(y_test,y_pred)))
```

```python
 # plotting the graph
plt.scatter(X_train,y_train,color="black")
plt.plot(X_train,y_pred_2,color="blue")
plt.xlabel("Hours")
plt.ylabel("Score")
plt.show()
```
